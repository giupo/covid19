```python
import pandas as pd
import requests

from IPython.display import HTML
import base64  

def create_download_link( df, title = "Download CSV file", filename = "DatiLombardia.csv"):  
    csv = df.to_csv(index =False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)
```

```python
url = "https://services1.arcgis.com/XannvQVnsM1hoZyv/arcgis/rest/services/INCR_DATE_PRV_TAMP_RL_v2/FeatureServer/0/query?f=json&where=1%3D1&returnGeometry=false&spatialRel=esriSpatialRelIntersects&outFields=ObjectId2%2CINCREMENTO%2CDATA_RICEVIMENTO_TAMPONE%2CPROVINCIA&orderByFields=DATA_RICEVIMENTO_TAMPONE%20asc&resultOffset=0&resultRecordCount=32000&resultType=standard&cacheHint=true"
res = requests.get(url)
data = res.json()
```

```python
data = data["features"]
```

```python
data = [x["attributes"] for x in data]
```

```python
data = pd.DataFrame(data)
```
